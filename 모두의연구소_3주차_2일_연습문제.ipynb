```python
# 문제 1
# 숫자를 하나 입력받아, 그 숫자가 5보다 큰지 확인하는 프로그램을 작성해보세요.

num = int(input("숫자를 입력하세요 :"))

if num == 5:
    print("입력하신 숫자는 5입니다.")
elif num > 5:
    print("입력하신 숫자가 5보다 큽니다.")
else:
    print("입력하신 순자가 5보다 작습니다.")


# 문제 2
# "안녕하세요!"를 5번 출력하는 프로그램을 작성해보세요.

for i in range(5):
    print("안녕하세요!")


# 문제3
# 1부터 10까지의 숫자 중에서, 짝수만 출력하는 프로그램을 작성해보세요.

for i in range(1, 11):
    if i % 2 == 0:
        print(i)

# 문제4
# 1부터 20까지의 숫자 중, 3의 배수만 출력하는 프로그램을 작성하세요.

for i in range(1, 21):
    if i % 3 == 0:
        print(i)

# 문제5
# 구구단 2단부터 9단까지 출력하는 프로그램을 작성하세요.

for i in range(2, 10):
    for j in range(1, 10):
        print(f"{i} * {j} = {i*j}")
    print()
```
