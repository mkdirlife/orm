# 1. 구글 코랩 AI

오늘은 구글 코랩 AI 사용법에 대해서 알아볼 거에요.

* colab 내에서 구글이 공개한 Codey 모델을 이용하여 사람의 명령(프롬프트 작성)
  으로 AI기반 코딩을 할 수 있음
* 코드 자동완성, 챗봇가능, Chat GPT와 동일한 방식으로 사용 가능
* Colab 내에서 언제든지 활용 가능
* 원래는 코랩 프로 유료 구독자들에게만 제공되었으나 얼마전에 무료로 기능을
  사용할 수 있게 되었음

```python
import random

# Create a 5x5 bingo board.
board = [[0 for _ in range(5)] for _ in range(5)]

# Fill the board with numbers from 1 to 25.
for i in range(5):
    for j in range(5):
        board[i][j] = random.randint(1, 25)

# Print the board.
for row in board:
    print(' '.join(map(str, row)))

# Get the user's numbers.
user_numbers = [int(x) for x in input('Enter your numbers: ').split()]

# Check if the user has won.
if all(num in user_numbers for num in board):
    print('You win!')
else:
    print('You lose.')
```
